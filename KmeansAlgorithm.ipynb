```python
def cal_distance(a,b):
    return (np.sum((a-b)**2))**0.5
```

```python
def nearest_Centroids(c, d):
    nearest_centroid = []
    for i in d:
        distance=[]
        for j in c:
            distance.append(cal_distance(i, j))
        nearest_centroid.append(np.argmin(distance))
        
    return nearest_centroid
```

```python
find_centroids = nearest_Centroids(centroid, Arr)
find_centroids
```

```python
def clus_centroids(clusters, Arr):
    nearest_centroids = []
    n_df = pd.concat([pd.DataFrame(Arr), pd.DataFrame(clusters, columns=['cluster'])],
                      axis=1)
    for k in set(n_df['cluster']):
        current_cluster = n_df[n_df['cluster'] == k][n_df.columns[:-1]]
        cluster_mean = current_cluster.mean(axis=0)
        nearest_centroids.append(cluster_mean)
    return nearest_centroids
```

```python
for i in range(10):
    find_centroids = nearest_Centroids(centroid, Arr)
    centroid = clus_centroids(find_centroids, Arr)
    
plt.figure()
plt.scatter(np.array(centroid)[:, 0], np.array(centroid)[:, 1], color='blue')
plt.scatter(Arr[:, 0], Arr[:, 1], alpha=0.1)
plt.show()
```

```python
plt.figure()
plt.scatter(np.array(centroid)[:, 0], np.array(centroid)[:, 1], color='red')
plt.scatter(Arr[:, 0], Arr[:, 1],c=find_centroids, cmap='magma' ,alpha=0.7)
plt.show()
```
